```python
import plotly.express as px
import pandas as pd


# Import data from USGS
data = pd.read_excel('/content/events_map.xlsx')


# Create scatter map
fig = px.scatter_geo(data, lat='Latitude', lon='Long', color='State',
                     hover_name='Event',
                     title='CF Licensed Events')
fig.show()
```

```python
fig = px.choropleth(data, locations='iso_alpha', hover_name='Event',
                    projection='natural earth', title='GDP per Capita by Country')
fig.show()
```
